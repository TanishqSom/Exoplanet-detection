```python
import pandas as pd
import numpy as np
```

```python
# Import dataset
df = pd.read_csv('exoplanets_2018.csv')
# Select top of the dataset
df.shape
```

```python
# Analysing type of data
df.info()
```

```python
# Importing columns_meaning data
col_meaning = pd.read_csv('columns_meaning.csv')
# Selecting top of col_meaning
col_meaning.head()
```

```python
# Cleaning col_meaning data
column = col_meaning.columns[0]
col_meaning = col_meaning[column].dropna()
col_meaning.head()
col_meaning = col_meaning[2:]
```

```python
# Renaming Columns for better understanding of dataset (optional)
renamed_col = {}
for name in col_meaning:
    new_name = name.replace('#','').replace(' ','').replace('COLUMN','').split(':')
    renamed_col[new_name[0]] = new_name[1]
df = df.rename(columns=renamed_col)
```

```python
# Viewing new dataset
df.head()
```

```python
# Creating numerical class of categorical data
df['ExoplanetCandidate'] = df['DispositionUsingKeplerData'].apply(lambda x: 1 if x == 'CANDIDATE' else 0)
df['ExoplanetConfirmed'] = df['ExoplanetArchiveDisposition'].apply(lambda x: 2 if x == 'CONFIRMED' else 1 if x == 'CANDIDATE' else 0 )
```

```python
# Viewing new dataset
df.head()
```

```python
# Removing unnecessary columns
df.drop(columns=['KeplerName',
                 'KOIName',
                 'EquilibriumTemperatureUpperUnc.[K]',
                 'KepID',
                 'ExoplanetArchiveDisposition',
                 'DispositionUsingKeplerData',
                 'NotTransit-LikeFalsePositiveFlag',
                 'StellarEclipseFalsePositiveFlag',
                 'CentroidOffsetFalsePositiveFlag',
                 'EphemerisMatchIndicatesContaminationFalsePositiveFlag',
                 'TCEDelivery',
                 'EquilibriumTemperatureLowerUnc.[K]'],
        inplace=True)
```

```python
#Checking which column has null values
df.isna().any()
```

```python
# Deleting Rows with null values
df.dropna(inplace=True)
# Checking Size on new dataset
df.shape
```

```python
# Preparing our data to feed our model
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

clean_dataset(df)
```

```python
# Saving cleaned data to csv
df.to_csv('DATA.csv',index=False)
```
